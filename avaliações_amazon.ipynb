```python
dados= pd.read_csv('/content/dados - dados.csv')
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados['actual_price']= dados['actual_price'].str.replace('₹','')
```

```python
dados['actual_price']= dados['actual_price'].str.replace('.','')
```

```python
dados['actual_price']= dados['actual_price'].str.replace(',','.')
```

```python
dados['product_price']= dados['product_price'].str.replace('₹','').str.replace('.','').str.replace(',','.')
```

```python
dados['rating']= dados['rating'].str.replace(',','.')
```

```python
dados['actual_price']= pd.to_numeric(dados['actual_price'])
dados['product_price']= pd.to_numeric(dados['product_price'])
dados['rating']= pd.to_numeric(dados['rating'])
dados.head()
```

```python
dados.info()
```

```python
dados.describe()
```

```python
dados.profile_report()
```

```python
avaliacoes_acima_de_45 = dados[dados['rating'] > 4.5]

# Use value_counts() para contar os valores únicos e suas contagens
contagem_avaliacoes = avaliacoes_acima_de_45['rating'].value_counts().sort_index()

# Crie um gráfico de barras para visualizar a contagem
plt.figure(figsize=(8, 6))
contagem_avaliacoes.plot(kind='bar')
plt.xlabel('Avaliação')
plt.ylabel('Contagem')
plt.title('Contagem de Avaliações Acima de 4.5')
plt.show()
```

```python
# Suponha que você tenha a taxa de câmbio em uma variável
taxa_de_cambio_inr_para_brl = 0.055  # Exemplo: 1 INR = 0.055 BRL

# Converta as colunas 'actual_price' e 'product_price' para reais
dados['actual_price_brl'] = dados['actual_price'] * taxa_de_cambio_inr_para_brl
dados['product_price_brl'] = dados['product_price'] * taxa_de_cambio_inr_para_brl
```

```python
dados.head()
```

```python
# Filtros
valores_ate_50 = dados[dados['actual_price_brl'] <= 50]
valores_50_a_500 = dados[(dados['actual_price_brl'] > 50) & (dados['actual_price_brl'] <= 500)]
valores_acima_de_500 = dados[dados['actual_price_brl'] > 500]
```

```python
contagem_ate_50 = valores_ate_50.shape[0]
contagem_50_a_500 = valores_50_a_500.shape[0]
contagem_acima_de_500 = valores_acima_de_500.shape[0]
```

```python
contagens = [contagem_ate_50, contagem_50_a_500, contagem_acima_de_500]
rotulos = ['Até R$50', 'De R$50 a R$500', 'Acima de R$500']
destacar = (0.1, 0, 0)  # Destaque o primeiro pedaço (Até R$50)

# Crie o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(contagens, labels=rotulos, explode=destacar, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Contagem de Produtos por Faixa de Valor')
plt.axis('equal')  # Mantenha o aspecto igual para que seja um círculo
plt.show()
```
